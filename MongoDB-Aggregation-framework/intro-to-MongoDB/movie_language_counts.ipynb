```python
# imports
from pymongo import MongoClient
import pprint
from IPython.display import clear_output
```

```python
# copy the URI from the cluster
client = MongoClient("""mongodb://analytics:analytics-password@mflix-shard-00-00-
4jt66.mongodb.net:27017,mflix-shard-00-01-4jt66.mongodb.net:27017,mflix-shard-00-
02-4jt66.mongodb.net:27017/mflix?ssl=true&replicaSet=mflix-shard-0&authSource=admin""")
print(client.mflix)
```

```python
# create a pipeline to pass to aggregate function, to count the languages
pipeline = [
    {
        '$group': {
            '_id': {"language": "$language"},
            'count': {"$sum": 1}
        }
    }
]
clear_output() # clears o/p in jupyter notebook
pprint.pprint(list(client.mflix.movies_initial.aggregate(pipeline)))
```

```python
# create a pipeline to pass to aggregate function, to sort the languages as per count
pipeline = [
    {
        '$group': {
            '_id': {"language": "$language"},
            'count': {"$sum": 1}
        }
    },
    {
        '$sort': {'count': -1}
    }
]
clear_output() # clears o/p in jupyter notebook
pprint.pprint(list(client.mflix.movies_initial.aggregate(pipeline)))
```
